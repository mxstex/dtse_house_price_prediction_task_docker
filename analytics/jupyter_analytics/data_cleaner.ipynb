```python
import pandas as pd
```

```python
input_data_path = "housing.csv"
EXPECTED_FEATURES = [
    'longitude', 'latitude', 'housing_median_age', 'total_rooms',
    'total_bedrooms', 'population', 'households', 'median_income',
    'ocean_proximity__LT_1H_OCEAN', 'ocean_proximity_INLAND',
    'ocean_proximity_ISLAND', 'ocean_proximity_NEAR_BAY',
    'ocean_proximity_NEAR_OCEAN'
]
```

```python
df =  pd.read_csv(input_data_path)
df
```

```python
# Step 2: Validate minimum columns
if df.shape[1] < 2:
    print(f"Insufficient columns in file: {input_data_path}")
else:
    print("File check OK")
```

```python
# Step 3: Normalize column names
df.columns = df.columns.str.lower()
print(f"Normalized column names: {list(df.columns)}")
df
```

```python
# Step 4: Rename columns for consistency
column_renames = {
    "lat": "latitude",
    "bedrooms": "total_bedrooms",
    "median_age": "housing_median_age",
    "pop": "population",
    "rooms": "total_rooms",
}
df.rename(columns=column_renames, inplace=True)
print(f"Renamed columns: {list(df.columns)}")
df
```

```python
# Step 5: Encode categorical variables
if "ocean_proximity" in df.columns:
    df = pd.get_dummies(df, columns=["ocean_proximity"], drop_first=False)
    print(f"Encoded 'ocean_proximity' column. Current columns: {list(df.columns)}")
else:
    print("'ocean_proximity' column not found. Skipping encoding.")
df
```

```python
# Step 6: Handle missing or unexpected values
df.replace("Null", 0, inplace=True)
df.fillna(0, inplace=True)
print(
    "Handled missing and unexpected values (replaced 'Null', filled NaNs)."
)
df
```

```python
# Step 7: Validate target column
target = "median_house_value"
if target not in df.columns:
    print(f"Target column '{target}' not found in the dataset.")
else:
    print(f"Target column found: {target}")
```

```python
# Step 8: Separate features and target; for passing data to model
y = df[target]
X = df.drop(columns=[target])
print(
    f"Separated target column '{target}'. Features shape: {X.shape}, Target shape: {y.shape}"
)
print(f"y: {y}")
print(f"X: {X}")
```

```python
# Step 9: Align features with the expected schema
for col in EXPECTED_FEATURES:
    if col not in X.columns:
        X[col] = 0  # Add missing columns as 0
        print(f"Added missing column '{col}' with default value 0.")
X = X[EXPECTED_FEATURES]
print(f"Aligned features with the expected schema. Final shape: {X.shape}")

print("Data preprocessing completed successfully.")
```
